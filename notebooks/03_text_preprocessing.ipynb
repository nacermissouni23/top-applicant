# 03 Text Preprocessing

## Purpose
Implementation of a spaCy-based NLP pipeline to transform raw text into a cleaned corpus optimized for data mining.

## NLP Pipeline
- Tokenization & Lemmatization.
- Stop-word & Special Character Removal.
- Language Detection (filtering for English or specific target languages).

## IO
- **Input**: Standardized text from `data/interim/`.
- **Output**: Preprocessed corpus stored in `data/processed/`.
