# 04 Feature Engineering

## Purpose
Extracting quantitative features from cleaned text, including TF-IDF vectorization, N-grams, and specialized entity extraction (e.g., skill tags).

## Mathematical Overview
The TF-IDF weight for a term $t$ in a document $d$ within a corpus $D$ is calculated as:

$$w_{t,d} = tf_{t,d} \cdot \log \frac{N}{df_t}$$

## IO
- **Input**: Preprocessed corpus from `data/processed/`.
- **Output**: Feature matrices and sparse vectors for machine learning models.
