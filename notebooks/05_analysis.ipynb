# 05 Analysis & Discovery

## Purpose
Aggregating features to identify skill clusters, role archetypes, and market trends. This is the primary research stage of the pipeline.

## Research Goals
- Identify recurring skill "motifs" in various job categories.
- Discover archetype clusters using unsupervised learning.
- Visualize frequency distributions and correlation matrices.

## IO
- **Input**: Engineered features and metadata.
- **Output**: Visualizations in `outputs/figures/` and tabular summaries in `outputs/tables/`.
