# 02 Data Cleaning

## Purpose
Loading raw data into pandas DataFrames, performing deduplication, and standardizing diverse schema formats into a unified structure.

## Cleaning Steps
- Deduplication: Identify and remove duplicate job postings based on unique IDs.
- Handling Missing Values: Impute or drop descriptions with insufficient content.
- Schema Mapping: Align different source fields (e.g., "title" vs "job_title") to a single standard.

## IO
- **Input**: Raw data from `data/raw/`.
- **Output**: Cleaned, standardized DataFrames stored in `data/interim/`.
